Load Model

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_id = "openai/gpt-oss-20b"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype="auto",
    device_map="cuda",
)
```

Function Generate Survey Response

```python
import re
import torch

def generate_survey_response(
    messages,
    model,
    tokenizer,
    max_new_tokens: int = 200,
    temperature: float = 1.2,
    top_p: float = 0.9,
    top_k: int = 40,
):
    """
    Generate a survey response and extract THREE numbers on a 1–7 scale from the final answer.

    Returns:
        numbers: list[int] of length 3
        decoded: full decoded text (for debugging)
        final_text: only the final channel content
    """

    # 1. Build model inputs with chat template
    inputs = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        return_tensors="pt",
        return_dict=True,
    ).to(model.device)

    # 2. Generate with sampling (for variability)
    with torch.no_grad():
        generated = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=True,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
        )

    # 3. Decode only the newly generated tokens
    decoded = tokenizer.decode(generated[0][inputs["input_ids"].shape[-1]:])

    # 4. Extract only the <channel>final<message> content (if present)
    final_text = decoded
    if "<|channel|>final" in final_text:
        final_text = final_text.split("<|channel|>final", 1)[1]

    # strip out <message> marker if present
    final_text = final_text.replace("<|message|>", "")

    # cut off anything after a new channel marker
    if "<|channel|>" in final_text:
        final_text = final_text.split("<|channel|>", 1)[0]

    final_text = final_text.strip()

    # 5. Find numbers between 1 and 7 in the final text
    nums = re.findall(r"\b[1-7]\b", final_text)

    # if not found there, fall back to full decoded text
    if len(nums) < 3:
        nums = re.findall(r"\b[1-7]\b", decoded)

    if len(nums) < 3:
        raise ValueError(
            "Could not find THREE scale numbers between 1 and 7.\n"
            f"Final text was:\n{final_text!r}\n\n"
            f"Decoded output was:\n{decoded!r}"
        )

    numbers = list(map(int, nums[:3]))
    return numbers, decoded, final_text
```

# Prompt (Paste here your prompt)

```python
import random
def sample_age():
    return random.randint(18, 35)

def sample_gender():
    return random.choice(["Male", "Female", "Non-binary"])

age = sample_age()
gender = sample_gender()
```

```python
messages = [
    {
        "role": "system",
        "content": "You are a market research participant responding to survey questions."
    },
    {"role": "user", "content": f"""You are {age} years old and identify as {gender}.
You have just viewed the Nike "Own the Floor" advertisement featuring a dancer and the slogan “OWN THE FLOOR”, emphasizing movement, empowerment, and self-expression.

Rate each statement on a 7-point scale (1 = completely disagree, 7 = completely agree).

Return ONLY three numbers separated by commas, no words.

Q1: I consider physical activity (including sports or dance) to be a central part of my lifestyle.
Q2: Wearing well-known athletic brands makes me feel more confident.
Q3: This advertisement makes me want to purchase Nike products.

Example format: 5,6,4"""
}
]
```

500 participants - 7 Likert

Preprocessing

```python
import pandas as pd
from tqdm.auto import tqdm
import os

# -----------------------------
# SET UP
# -----------------------------

N = 500                           # Total number of responses
autosave_every = 10                # Save each 10 responses
output_file = "/content/drive/MyDrive/market_intelligence_survey_results_500.csv" #YOU WRITE THE NAME OF THE CSV FILE

# -----------------------------
# LOAD FILE IF EXISTS
# -----------------------------

if os.path.exists(output_file):
    df_existing = pd.read_csv(output_file)
    responses = df_existing[["Age", "Gender", "Q1", "Q2", "Q3"]].to_dict("records")
    start_idx = len(responses)
    print(f"Found existing file with {start_idx} responses. Resuming…")
else:
    responses = []
    start_idx = 0
    print("No existing file found. Starting fresh.")
```

Main Loop

```python
# -----------------------------
# MAIN LOOP
# -----------------------------

for i in tqdm(range(start_idx, N), desc="Simulating respondents"):
    age = sample_age()
    gender = sample_gender()
    numbers, _, _ = generate_survey_response(
        messages,
        model,
        tokenizer,
        max_new_tokens=200,   # necessary for model to reach <final>
        temperature=1.2,
        top_p=0.9,
        top_k=40,
    )
    q1, q2, q3 = numbers
    responses.append({
    "Age": age,
    "Gender": gender,
    "Q1": q1,
    "Q2": q2,
    "Q3": q3
})

    # Peridical save
    if (i + 1) % autosave_every == 0:
        df_partial = pd.DataFrame(responses, columns=["Age", "Gender", "Q1", "Q2", "Q3"])
        df_partial.to_csv(output_file, index=False)
        print(f"Autosaved {i + 1} responses to {output_file}")
```
